I can think of a couple of columns to keep for example: `title`, `country`, `length`, `publicationDate`, apprently I could have used this in read_csv step (which could lead to lower memory usage and therefore be used in the songs dataset)

```python
columns_to_keep = ['title', 'country', 'length', 'publicationDate']
albums_df = pd.read_csv('/kaggle/input/wasabi-song-corpus/wasabi_albums.csv', usecols=columns_to_keep)
display(albums_df)
```

To check whether our original idea if feasable or not we'll need to have many countries and many non null values for this country field, therefore I'll group by it to have a general idea of the data I have.

```python
# Group by 'country' and count the number of songs for each country
country_counts = albums_df.groupby('country').size()

# Convert the result to a DataFrame for better visualization
country_counts_df = country_counts.reset_index(name='number_of_songs')
country_counts_df = country_counts_df.sort_values(by='number_of_songs', ascending=False)
display(country_counts_df)
```

Therefore the idea is doable. I'll now continue with the pre-processing.

```python
albums_df_cleaned = albums_df.dropna(subset=['title', 'country'])
display(albums_df_cleaned)
```
